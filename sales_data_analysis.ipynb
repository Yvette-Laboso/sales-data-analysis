```python
import pandas as pd
df= pd.read_csv('SampleSuperstore.csv.zip')
df.head()
```

```python
df.info()
```

```python
df.groupby('Category')['Sales'].sum()
```

```python
import matplotlib.pyplot as plt
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

Technology generated the highest sales compared to furniture and office supplies

```python
df.groupby('Region')['Profit'].sum()
```

```python
df.groupby('Region')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Region")
plt.xlabel("Profit")
plt.ylabel("Region")
plt.show()
```

The West Region makes the most profit
